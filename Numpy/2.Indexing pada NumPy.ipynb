## Index array 2D (matrix)

```python
arr_2d = np.array(([5,10,15],
                   [20,25,30],
                   [35,40,45]))

arr_2d
```

```python
arr = np.array([1,2,3])
arr
```

```python
arr[0]
```

```python
arr_2d[1]
```

```python
# Formatnya adalah arr_2d[baris][kolom] atau arr_2d[baris,kolom]

arr_2d[1][0]
```

```python
arr_2d[1,0]
```

```python
arr_2d
```

```python
# Slicing pada array 2D

arr_2d[:2,1:]
```

```python
arr_2d[2]
```

```python
arr_2d[2,-1]
```
